```python
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
import os
import time
```

```python
def init_driver():
    driver = webdriver.Edge()
    return driver

def get_bing_image_urls(query, num_images=50):
    driver = init_driver()
    url = f"https://www.bing.com/images/search?q={query}"
    driver.get(url)

    image_urls = set()
    scrolls = 0
    while len(image_urls) < num_images and scrolls < 200:  # 限制滚动次数
        # 查找所有缩略图
        thumbnails = driver.find_elements(By.XPATH, "//a[@class='iusc']")
        for thumbnail in thumbnails:
            try:
                metadata = thumbnail.get_attribute("m")
                if metadata:
                    image_url = eval(metadata).get("murl")  # 提取高清图片链接
                    if image_url and image_url not in image_urls:
                        image_urls.add(image_url)

                # 如果达到了所需的图片数量，则停止
                if len(image_urls) >= num_images:
                    break
            except Exception as e:
                print(f"解析图片链接时出错: {e}")

        # 尝试点击 "查看更多图片" 按钮
        try:
            see_more_button = driver.find_element(By.XPATH, "//a[contains(@class, 'btn_seemore')]")
            if see_more_button.is_displayed():
                print("点击查看更多图片...")
                driver.execute_script("arguments[0].click();", see_more_button)
                time.sleep(3)  # 等待加载新图片
        except (NoSuchElementException, ElementClickInterceptedException):
            print("未找到查看更多按钮或无法点击")

        # 滚动页面
        driver.execute_script("window.scrollBy(0, document.body.scrollHeight);")
        time.sleep(3)  # 增加加载时间
        scrolls += 1

    driver.quit()
    return list(image_urls)

def download_images(image_urls, query, classes):
    folder=f"{query}"
    if not os.path.exists(folder):
        os.makedirs(folder)

    file_index = 1
    for url in image_urls:
        try:
            response = requests.get(url, timeout=10)
            if response.status_code == 200:
                # ext = url.split(".")[-1]
                # if len(ext) > 4:
                #     ext = "jpg"
                file_name = os.path.join(folder, f"{classes:03d}_2_{file_index:04d}.jpg")
                with open(file_name, "wb") as file:
                    file.write(response.content)
                print(f"图片已保存: {file_name}")
                file_index += 1  # 确保下一个文件名是连续的
        except Exception as e:
            print(f"下载图片失败: {e}")



```

```python
if __name__ == "__main__":
    query = "stop check traffic sign"  # 搜索关键词
    classes = 57
    num_images = 300  # 下载图片数量
    image_urls = get_bing_image_urls(query, num_images)
    print(f"获取到 {len(image_urls)} 张图片链接。")

    download_images(image_urls,query,classes)
```
